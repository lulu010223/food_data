```python
def program1():

    import difflib                      # 추천 프로그램 실행에서 제품 이름이 가장 유사한 항목을 찾기 위해 사용

    # 데이터를 읽어서 리스트로 저장
    data = []
    with open(directory_path + 'food_data.csv', 'r') as f:
        reader = csv.reader(f)
        header = next(reader)  # 헤더 건너뛰기
        for row in reader:
            data.append(row)

    # 데이터를 numpy 배열로 변환
    data_array = np.array(data)

    # csv 파일의 데이터를 기반으로 pandas DataFrame을 생성
    data_frame = pd.DataFrame(data, columns = header)
    # 인덱스를 '제품 번호'의 데이터로 변경하기
    data_frame = data_frame.set_index(keys = '제품 번호')
    # 제품을 검색해서 결과가 존재하면 저장할 변수
    searched_name = ''


    # 프로그램을 종료하지 않는 이상 계속 반복해서 검색을 할 수 있도록한다.
    while True:
        # 제품 이름 검색
        print("\n🟰🟰🟰🟰🟰🟰🟰🟰🟰🔎검색🟰🟰🟰🟰🟰🟰🟰🟰🟰")
        print("제품 이름을 입력하세요 (종료하려면 'X' 입력)\n")
        name = input('🔍: ')

        # 'X'를 누르면 프로그램을 종료할 수 있도록 한다.
        if name == 'X':
            print("검색을 종료합니다.")
            break

        print('\n검색 중입니다... ⏳\n')

        # 데이터 프레임의 '제품 이름' 열과 검색어를 비교하여 검색어 찾고, 검색어가 존재한다면 search_result에 저장
        # food_data.csv 파일의 제품 이름 데이터에서 공백을 제거하여, 공백이 포함된 검색어와도 일치한다.
        search_result = data_frame[data_frame['제품 이름'].str.replace(" ", "") == name.replace(" ", "")]

        # 검색어가 없는 경우 추천 프로그램 실행
        if search_result.empty:
          print('\n>>검색어와 일치하는 결과가 없습니다. 현재 검색어와 가장 유사한 검색어를 보여드립니다.')

          # 기존 결과가 없는 경우
          # 데이터프레임에 유사도 더하여 가장 유사한 이름 찾기
          # difflib.SequenceMatcher의 ratio() 메서드는 두 문자열 간의 유사도를 0에서 1 사이의 값으로 반환하고, 1에 가까울수록 유사도가 높다.
          # 람다 함수로 데이터 프레임의 각 행의 제품 이름과 입력된 이름 간의 유사도를 계산 (None은 무시할 문자, x는 제품 이름, name은 검색어)
          similarity = data_frame['제품 이름'].apply(lambda x: difflib.SequenceMatcher(None, x, name).ratio())
          # 유사도가 가장 높은 품목을 저장
          max_similarity_index = similarity.idxmax()
          # x.loc[]를 사용해 인덱스 이름으로 유사도가 가장 높은 품목의 이름을 저장하기
          max_similarity_name = data_frame.loc[max_similarity_index, '제품 이름']
          print("\n🍴이름이 가장 유사한 품목:", max_similarity_name, "\n\n")

          # 기존 결과가 있는 경우
          if searched_name != '':
            print('\n>>기존에 검색한 결과와 유사한 식품입니다.')
            # 기존 결과의 검색어와 이름이 가장 비슷한 품목을 찾기
            # 데이터 프레임의 '제품 이름' 열과 입력된 이름 간의 유사도 계산
            similarity = data_frame['제품 이름'].apply(lambda x: difflib.SequenceMatcher(None, x, searched_name).ratio())
            # sort_values를 사용해 유사도 순서로 정렬한 인덱스
            sorted_index = similarity.sort_values(ascending=False).index
            # 기존 검색어를 제외하고 유사한 품목의 이름
            second_similar_name = data_frame.loc[sorted_index[1], '제품 이름']
            # 기존 검색어와 동일한 식품 분류의 품목
            same_category_items = data_frame[data_frame['식품 분류'] == category]
            same_alternative_items = data_frame[data_frame['감미료 정보'] == sweetener_info]
            print("\n🍴이름이 가장 유사한 품목:", second_similar_name)

            #KeyError가 일어나지 않도록 식품 분류가 존재하는지 검토
            if not same_category_items.empty:
              # 데이터프레임에서 한 가지를 랜덤으로
              random_category_item = same_category_items.sample(n=1)['제품 이름'].values[0]
              # 기존 검색어와 random_category_item의 결과가 같다면 다시 실행하여 결과값 바꾸기
              # 그러나 '식품 유형'에 해당 데이터 밖에 존재하지 않는다면 값을 바꾸지 않기
              while len(same_category_items) > 1 and random_category_item == searched_name:
                random_category_item = same_category_items.sample(n=1)['제품 이름'].values[0]
              # 같지 않다면 출력
              print("🍴같은 식품 분류의 품목:", random_category_item)

            #KeyError가 일어나지 않도록 감미료 정보가 존재하는지 검토
            if not same_alternative_items.empty:
              # 데이터프레임에서 한 가지를 랜덤으로
              random_alternative_item = same_alternative_items.sample(n=1)['제품 이름'].values[0]
              # 기존 검색어와 random_alternative_item의 결과가 같다면 다시 실행하여 결과값 바꾸기
              # 그러나 '감미료 정보'에 해당 데이터 밖에 존재하지 않는다면 값을 바꾸지 않기
              while len(same_alternative_items) > 1 and random_alternative_item == searched_name:
                random_alternative_item = same_alternative_items.sample(n=1)['제품 이름'].values[0]
              # 같지 않다면 출력
              print("🍴같은 감미료를 사용한 품목:", random_alternative_item, "\n\n")

        # 결과가 있는 경우
        else:
            # 검색된 제품의 이름을 search_name에 저장
            search_name = search_result['제품 이름'].values[0]
            # 검색된 제품의 단위 정보를 gram_info에 저장
            gram_info = search_result['g당'].values[0]
            # 검색된 제품의 총 열량 정보를 total_calories에 저장
            total_calories = search_result['총 열량'].values[0]
            # 검색된 제품의 총 당류 정보를 total_sugar에 저장
            total_sugar = search_result['총 당류'].values[0]
            # 검색된 제품의 감미료 정보를 sweetener_info에 저장
            sweetener_info = search_result['감미료 정보'].values[0]
            # 검색된 제품의 식품 분류를 category에 저장
            category = search_result['식품 분류'].values[0]

            # 기존 검색어와 유사한 제품 추천에 사용하기 위한 변수
            # 기존 결과의 검색어
            searched_name = name

            # 검색한 제품의 총 열량, 총 당류, 사용된 감미료 정보를 출력한다.
            print("[", search_name, "] 의 총 열량:", gram_info, "당", total_calories, "kcal")
            print("[", search_name, "] 의 총 당류:", gram_info, "당", total_sugar, "g")
            print("[", search_name, "] 에 사용된 감미료:", sweetener_info, "\n")

            # 폰트 지정
            plt.rc('font', family='NanumBarunGothic')
            plt.rcParams['axes.unicode_minus'] = False

            # 결과값 막대 그래프로 표현하기
            x = ['총 열량', '총 당류']
            y = [float(total_calories), float(total_sugar)]

            # 막대 그래프 형식
            plt.figure(figsize=(4.5, 4.5))
            plt.style.use('ggplot')
            plt.bar(x, y)
            plt.title(search_name)
            plt.xlabel('항목')
            plt.show()

            # 결과값 파이차트로 표현하기
            kcal_values = [float(total_calories), 2000 - float(total_calories)]
            sweet_values = [float(total_sugar), 100 - float(total_sugar)]
            label = ['함유량', '기준치 - 함유량']
            color = ['r', 'gray']

            # 칼로리가 0 이상인 경우 파이차트 출력
            if float(total_calories) > 0:
                plt.pie(kcal_values, colors=color, autopct='%.1f%%', explode=[0.1, 0])
                plt.title('칼로리 1일 영양성분 기준치(2000kcal)에\n따른 함유량')
                plt.legend(label)
                plt.show()

            # 당류가 0 이상인 경우 파이차트 출력
            if float(total_sugar) > 0:
                plt.pie(sweet_values, colors=color, autopct='%.1f%%', explode=[0.1, 0])
                plt.title('당류 1일 영양성분 기준치(100g)에\n따른 함유량')
                plt.legend(label)
                plt.show()

def program2():

    val_list = a.values.tolist()

    while True :
      x = input("\n🟰🟰🟰🟰 원하시는 식품 분류 코드(알파벳)를 입력하세요🍴 🟰🟰🟰🟰\n\nA : 🥐과자류, 빵류 또는 떡류\nB : 🥣농산가공식품류\nC : 🍦빙과류\nD : 🧃음료류\nE : 🏺잼류\nF : 🧂조미식품류\nG : 🍺주류\nH : 🍱즉석섭취식품류\nI : 🍫코코아 가공품, 초콜릿류\n\n❌종료를 원하시면 'X' 입력\n\n")
      if x == 'X' :
        print("카테고리별 식품 리스트 보기 기능을 종료합니다.\n🟰🟰🟰🟰🟰🟰🟰🟰🟰🟰🟰🟰🟰🟰🟰🟰🟰🟰🟰🟰🟰🟰🟰🟰\n\n")
        break
      elif x == 'A' :
        print('\n💟 과자류, 빵류 또는 떡류에 해당하는 식품 목록은 아래와 같습니다👇\n')
        for i in val_list[0][1] :
          print(i)
        print('\n')
      elif x == 'B' :
        print('\n💟 농산가공식품류에 해당하는 식품 목록은 아래와 같습니다👇\n')
        for i in val_list[1][1] :
          print(i)
        print('\n')
      elif x == 'C' :
        print('\n💟 빙과류에 해당하는 식품 목록은 아래와 같습니다👇\n')
        for i in val_list[2][1] :
            print(i)
        print('\n')
      elif x == 'D' :
        print('\n💟 음료류에 해당하는 식품 목록은 아래와 같습니다👇\n')
        for i in val_list[3][1] :
          print(i)
        print('\n')
      elif x == 'E' :
        print('\n💟 잼류에 해당하는 식품 목록은 아래와 같습니다👇\n')
        for i in val_list[4][1] :
          print(i)
        print('\n')
      elif x == 'F' :
        print('\n💟 조미식품류에 해당하는 식품 목록은 아래와 같습니다👇\n')
        for i in val_list[5][1] :
          print(i)
        print('\n')
      elif x == 'G' :
        print('\n💟 주류에 해당하는 식품 목록은 아래와 같습니다👇\n')
        for i in val_list[6][1] :
          print(i)
        print('\n')
      elif x == 'H' :
        print('\n💟 즉석섭취식품류에 해당하는 식품 목록은 아래와 같습니다👇\n')
        for i in val_list[7][1] :
          print(i)
        print('\n')
      elif x == 'I' :
        print('\n💟 코코아 가공품, 초콜릿류에 해당하는 식품 목록은 아래와 같습니다👇\n')
        for i in val_list[8][1] :
            print(i)
        print('\n')
      else:
        print('\n⚠️ 다시입력해주세요\n')
        print("2. 카테고리 별 식품 리스트 보기")

def program3():
    data = np.genfromtxt(directory_path + '감미료_정보.csv', delimiter=',', dtype=str)
    header = data[0]
    data_array = data[1:]
    #주어진 감미료들 중 하나를 검색
    print('''
🍬🍬🍬감미료 15가지 중 하나를 검색해보세요🍬🍬🍬
ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ
아스파탐, 수크랄로스, 알룰로스, 에리스리톨, 아세설팜칼륨, D-소르비톨액, 효소처리스테이바,
말티톨, 자일리톨, 비정제원당, 폴리글리시톨 시럽, 나한과 추출물, 토마틴, 스테비올배당체, 당알콜
ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ
    ''')
    #궁금한 감미료 검색
    while True:
        sugar_name = input('궁금한 감미료를 입력해주세요(프로그램 종료 시 X 입력): ')  #프로그램 종료 기능
        if sugar_name == 'X':
            print("프로그램을 종료합니다.")
            break

        sugar_info = False
        #감미료의 장점 및 부작용을 저장할 변수
        sugar = []

        for row in data_array:
            if sugar_name in row[1]:           #in 연산자를 활용해 해당 감미료 찾기
                sugar = row[2:].tolist()
                #궁금한 감미료에 대한 정보를 출력하도록 한다
                print(f"[{sugar_name}], 의 장점과 부작용은 다음과 같습니다:\n{sugar}")
                sugar_info = True
                break

        if not sugar_info:
            print('정확한 감미료 명칭을 입력해주세요.')  #검색한 결과가 없을 때 다시 재검색할 수 있도록 한다


def program4():

    def search_food_with_alternatives(df, search_term):
        # 검색어를 포함하는 행 필터링
        filtered_df = df[df['감미료 정보'].str.contains(search_term)]

        # 대체당이 있는 식품 필터링
        filtered_df = filtered_df[filtered_df['감미료 정보'].notnull()]

        # 식품 리스트 반환
        food_list = filtered_df['제품 이름'].tolist()
        return food_list

    file_path = '/content/gdrive/My Drive/파데분_11조/food_data.csv'

    sweeteners = [
    '감초추출물', '글리실리진산이나트륨', '네오탐', '락티톨', 'D-리보오스', '만니톨', 'D-말티톨',
    '말티톨시럽', '사카린나트륨', 'D-소비톨', 'D-소비톨액', '수크랄로스', '스테비올배당체', '아세설팔칼륨',
    '아스파탐', '에리스리톨', '이소말트', 'D-자일로오스', '자일리톨', '토마틴', '폴리글리시톨시럽',
    '효소처리스테비아', '알룰로스'
    ]

    print('''
✔식약처에서 승인한 감미료 22개는 다음과 같습니다✔
ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ
감초추출물, 글리실리진산이나트륨, 네오탐, 락티톨, D-리보오스, 만니톨, D-말티톨, 말티톨시럽
사카린나트륨, D-소비톨, D-소비톨액, 수크랄로스, 스테비올배당체, 아세설팔칼륨, 아스파탐
에리스리톨, 이소말트, D-자일로오스, 자일리톨, 토마틴, 폴리글리시톨시럽, 효소처리스테비아
ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ
추가로 대중적으로 알려진 알룰로스는 식약처에서 감미료가 아닌 '식품원료'로 구분되어 있습니다.
감미료 22가지와 알룰로스 중에서 검색해주세요! ╰(*°▽°*)╯
X를 입력하면 종료됩니다.
    ''')

    # csv 파일 읽기
    df = pd.read_csv(file_path)

    while True:
        search_term = input("\n찾으시는 감미료를 입력해주세요 : ")

        # 사용자가 X를 입력하면 종료
        if search_term == 'X':
            break
        elif search_term in sweeteners:
            food_list = search_food_with_alternatives(df, search_term)

            # 검색한 감미료가 포함된 식품이 있을 때
            if food_list:
                print('''
이 감미료를 사용한 식품들은 다음과 같습니다>>
                ''')
                print(food_list)

            # 검색한 감미료가 포함된 식품이 없을 때
            else:
                print('''
해당 감미료가 사용된 식품이 없습니다.
                ''')

        # sweeteners에 없는 감미료를 검색했을 때 재검색 알림
        else:
            print("리스트에 없는 감미료입니다. 목록을 확인하신 후 다시 입력해 주세요.\n")

# 프로그램을 선택하는 메뉴를 표시하고, 선택된 프로그램을 실행
while True:
    print("🍽️ 메뉴 선택하기\n")
    print("1. 제품 검색으로 당류와 감미료 정보 찾기 & 제품 추천")
    print("2. 카테고리별 식품 리스트 보기")
    print("3. 감미료별 장단점 검색하기")
    print("4. 감미료별 식품 리스트 검색하기")
    print("5. 종료")

    choice = input("\n실행하고자 하는 메뉴를 선택해주세요(숫자입력): ")

    if choice == "1":
        program1()
    elif choice == "2":
        program2()
    elif choice == "3":
        program3()
    elif choice == "4":
        program4()
    elif choice == "5":
        print("프로그램을 종료합니다.\n")
        break
    else:
        print("올바른 숫자를 입력해주세요.\n")
```
